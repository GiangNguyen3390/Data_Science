<h2 style="color:red"> 4. Boolean Indexing </h2>

<h3 style="color:blue"> We can Boolean-Index a pandas Series.</h3>

````python
ser = pd.Series(np.random.randn(20))

print("This series of random numbers has:", len(ser), 'items.')
print("--"*30)

print(ser)
````

````python
condition = ser > 0
print(condition)
````

````python
sub_ser = ser[condition]

print('The "greater than 0" sub-series of "ser" has:', len(sub_ser), 'items.')
print("--"*30)

print(sub_ser.head())
````

<h2 style="color:red"> 5. Grouping by Categorical Features. </h2>

**This will group the rows of your df by a categorical feature, This is really useful for answering qquestions like:**
> What is the average number of males accross all counties in California ---- > 329088.5

> What is the average population accross all Counties in Alabama --- > 72,098

````python
cs.head()
````

````python
by_state_max = cs.groupby("State").mean().head()
by_state_max.head()
````

<br>

````python
by_state_max.sort_values(by="TotalPop", ascending=False)
````

````python
cali = cs[cs['State'] == "California"]
cali.sort_values(by="TotalPop", ascending=False)
````
